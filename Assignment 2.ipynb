```python
# Unique Words in a Sentence
sentence = input("Enter sentence: ")
words_list = sentence.split()
unique_words_set = set(words_list)
print("Unique words count:", len(unique_words_set))
print("Unique words:", unique_words_set)
```

```python
#Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_paid_employee = None
highest_salary = 0
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_paid_employee = name
print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

```python
#Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
maximum_value = numbers[0]
minimum_value = numbers[0]
for value in numbers:
    if value > maximum_value:
        maximum_value = value
    if value < minimum_value:
        minimum_value = value
print("List:", numbers)
print("Max:", maximum_value)
print("Min:", minimum_value)
```

```python
#Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count_above_limit = 0
for price in prices:
    if price > 1000:
        count_above_limit += 1
print("Products above 1000:", count_above_limit)
```

```python
#Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
present_days = 0
for status in attendance:
    if status == "P":
        present_days += 1
attendance_percentage = (present_days / len(attendance))*100
print("Attendance Percentage:", attendance_percentage)
```

```python
#Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_phone_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_phone_numbers)
```

```python
#Count Character Frequency
text = "python"
character_count = {}
for char in text:
    if char in character_count:
        character_count[char] += 1
    else:
        character_count[char] = 1
print(character_count)
```

```python
#Check if a Key Exists in a Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000
}
if "Ravi" in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#Calculate Average Marks
marks = [70, 75, 80, 65]
total_marks = 0
for score in marks:
    total_marks += score
average_marks = total_marks / len(marks)
print("Average Marks:", average_marks)
```
